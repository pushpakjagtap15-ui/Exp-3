```python
# Grocery Shopping List
ShoppingList = []

def ItemAdd():
    # Adding items to the list using append()
    Item = input("Enter the name of the item > ")
    ShoppingList.append(Item)

    # Asking user for continuation
    Decision = input("Do you want to add more items? (Y/N) > ").strip().upper()

    if Decision == "Y":
        ItemAdd()
    else:
        # Displaying the final formatted list
        print("\n******* Your Final Shopping List **********")
        if not ShoppingList:
            print("The list is empty.")
        else:
            # Using enumerate to handle the 'Numbering' logic
            for index, item in enumerate(ShoppingList, start=1):
                print(f"{index}. {item}")

# Execute the function
ItemAdd()
```
